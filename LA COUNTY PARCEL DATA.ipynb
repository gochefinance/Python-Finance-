## GETTING PARCEL DATA FROM PUBLIC RECORDS

```python
import json
file=open(r"C:\Users\sespejog\OneDrive\Data Sets\LA COUNTY DATA\90031_Rollyear_2018_2019.json")
dictionary=json.load(file)
file.close()
```

Our Json file is neatly seperated into two parts "meta" and "data". Our "meta" part tells us more information about the data itself and our "data" part is where our actual parcel data will be. 

```python
dictionary.keys()
```

It is important to read our JSON data because it tells us what our data means. Los Angeles county might classify its parcel data different than other counties so we should look at our data before we make a decision.

```python
dictionary["meta"]
```

After analyzing the metadata we can see that if the value for the "id" key is equal to -1, it is a meta dictionary. Otherwise it tells us information about our columns. We are going to use this to extract information about our columns using conditional statements.

```python

for i in dictionary["meta"]["view"]['columns']:
    if i["id"]!= -1:
        try:
            print("*",i['name'],'----------',i["description"],"\n")
        except KeyError:
            pass
```

Now that we have our data we can create our own database and add or delete the columns we do not need.

```python
import pandas as pd
len(dictionary["meta"]["view"]['columns'])
col=[i["name"] for i in dictionary["meta"]["view"]['columns']]
real_estate= pd.DataFrame(dictionary["data"],columns=col)
real_estate.head()
```

We still haven't deleted the columns that have meta information. We are going to do this by checking the columns id. (The one that is equal to -1 if they are a meta column).

```python
columns=[]## we store the columns we want to store here
for i in dictionary["meta"]["view"]['columns']:
    if i["id"]!= -1:
        try:
            columns.append(i['name'])
        except KeyError:
            pass
print("These are the rows and columns BEFORE we got rid of metadata:",real_estate.shape)
real_estate=real_estate.copy() # make sure to create a copy of the dataframe before you edit columns out
for i in real_estate.columns:
    if i not in columns:
        real_estate=real_estate.drop(i,axis=1)#here we drop the columns
print("These are the rows and columns AFTER we got rid of metadata:",real_estate.shape)
```

Now that we ha have our dataframe we can take a random sample to make sure everything looks right.

```python
pd.set_option('display.max_columns', 999) #this makes sure we can see all our columns
real_estate.sample(5)
```
