## The Trade War

The plot above shows how Alibaba has been consistently showing promising returns in its Yearly reports, except for the time period after 2018 where Alibaba's stock price drops drastically. This is not because it started underperfoming but because of investor sentiment toward the US and China trade war relations.

This should not concern Alibaba since most of its business revolves in China so tariffs would not have a severe impact in its business model. However, Alibaba has been severely impacted everytime US and China Trade War relations escalade. The graph below shows Alibaba's stock price and it marks all the dates in which the US and China did something to heighten this conflict.


```python
important_dates=["2017-3-31","2017-7-19",
                 "2017-8-14","2018-1-17",
                "2018-1-22","2018-3-8",
                 "2018-4-2","2018-4-4",
                 "2018-6-15","2018-7-10",
                 "2018-8-1","2018-8-7",
                 "2018-8-23","2018-9-7",
                 "2018-9-24","2019-3-8",
                 "2019-8-1","2019-8-5",
                 "2019-8-9"
                ]
baba_start=start_date(2017,2,28)
Trade_War_BABA=load_stock("BABA",baba_start)
fig = plt.figure(figsize=(20, 7))
ax=Trade_War_BABA["Adj Close BABA"].plot()
ymin, ymax = ax.get_ylim()
#BABA_20F_dates=["2015-3-31","2016-3-31","2017-3-31","2018-3-31","2019-3-31"]
ax.vlines(x=important_dates, ymin=ymin, ymax=ymax-1, color='r')
ax.vlines(x=BABA_20F_dates, ymin=ymin, ymax=ymax-1, color='b')
```

For the first half of the trade war, Alibaba's stock seems to not react to any negative news regarding the trade war, however as the trade war escalated, Alibaba's stock started to react more negatively to bad news regarding the trade war. To see the effect of the escalated tensions of the trade war regarding stock price we are going to just focus on the later half of the trade war.

```python
important_dates=["2017-3-31","2017-7-19",
                 "2017-8-14","2018-1-17",
                "2018-1-22","2018-3-8",
                 "2018-4-2","2018-4-4",
                 "2018-6-15","2018-7-10",
                 "2018-8-1","2018-8-7",
                 "2018-8-23","2018-9-7",
                 "2018-9-24","2019-3-8",
                 "2019-8-1","2019-8-5",
                 "2019-8-9"
                ]
quarterly_reports=["2019-8-14","2019-5-14",
                  "2019-2-14","2018-11-14",
                   "2018-8-14",
                  "2018-5-14","2018-2-14",
                  "2017-11-14","2017-8-14",
                  "2017-5-14","2017-2-14"
                  ]
baba_start=start_date(2018,4,1)
Trade_War_BABA=load_stock("BABA",baba_start)
fig = plt.figure(figsize=(20, 7))
ax=Trade_War_BABA["Adj Close BABA"].plot()
ymin, ymax = ax.get_ylim()
#BABA_20F_dates=["2015-3-31","2016-3-31","2017-3-31","2018-3-31","2019-3-31"]
ax.vlines(x=important_dates, ymin=ymin, ymax=ymax-1, color='r')
ax.vlines(x=BABA_20F_dates, ymin=ymin, ymax=ymax-1, color='b')
ax.vlines(x=quarterly_reports, ymin=ymin, ymax=ymax-1, color='g')
plt.legend(("Stock Price", 'Bad Trade War News','Yearly 20-F Reports','Quarterly Earning Reports' ),
           loc='upper right')
plt.show()
```

As you can see any news regarding the trade war sends the stock tumbling down and the price of the stock still has not recovered.
However, good earning reports seem to break the spell that the Trade War has in investor eyes. Tomorrow we will see if their earning's report is positive enough to break the curse.
